```python
# ejmeplo importar pandas

import pandas as pd # importa la libreria pandas y le coloco el apodo pd
from sklearn.datasets import load_iris # importa la funcion load_iris de la libreria sklearn
```

```python
# Crear un datafram a partir de un diccionario

datos_inical = {"edad": [18, 18, 25, 32],
                "estatura": [1.7, 1.65, 1.82, 2],
                "nacionalidad": ["Peru", "Peru", "EEUU", "Ucrania"],
                "c_hijos": [1, "dos", "cero", 2] }

df = pd.DataFrame(datos_inical) # crea un dataframe a partir del diccionario datos_inical
```

```python
df
```

```python
df.info() # muestra informacion del dataframe df
```

```python
df.describe() # muestra estadisticas descriptivas del dataframe df
```

```python
# Generar grafico de barras
df["edad"].plot(kind="bar") # genera un grafico de barras de la columna edad del dataframe df
```

```python
df["nacionalidad"].plot(kind="pie") # genera un grafico de barras de la columna nacionalidad del dataframe df
```

```python
frecuancias = df["nacionalidad"].value_counts() # cuenta las frecuencias de la columna nacionalidad del dataframe df
frecuancias.plot(kind="pie", legend=frecuancias.all(), autopct='%1.1f%%') # genera un grafico de barras de las frecuencias de la columna nacionalidad del dataframe df
```

```python
# graficar una nube de puntos 
df.plot(kind="scatter", x="edad", y="estatura") # genera un grafico de barras de la columna edad vs estatura del dataframe df
```

```python
# guardar mi df en un archivo csv
df.to_csv("df_23_08.csv", index=False) # guarda el dataframe df en un archivo
df.to_excel("df_23_08.xlsx", index=False) # guarda el dataframe df en un archivo excel
df.to_json("df_23_08.json", index=False) # guarda el dataframe df
```
